### Your answer to c)

> As we have established already, tuples are immutible and cannot be changed. But however, if a tuple contains a list or dictionary, the tuple cannot be changed, but the contents of the objects inside the tuple can still be modified. So you cannot change the structure of a tuple since a tuple is a static arrangement of items, but the items inside the tuple can be modified since it does not change the tuples arrangement/structure.

### Your answer to d)

````python
# Answer to d)

set_1 = {1, 2, 3, 4, 5, 6}
set_2 = {4, 5, 6, 7, 8, 9}

# Union first method
set_union = set_1.union(set_2)
print(f"Union without changing first set:\n{set_union}")

# Union second method
set_1.update(set_2)
print(f"Union by changing the first set:\n{set_1}")

# Intersection first method
set_intersection = set_1.intersection(set_2)
print(f"Intersection without changing first set:\n{set_intersection}")

# Intersection second method
set_1.intersection_update(set_2)
print(f"Intersection by changing the first set:\n{set_1}")

set_1 = {1, 2, 3, 4, 5, 6} # Reset

# Difference first method
set_difference = set_1.difference(set_2)
print(f"Difference without changing first set:\n{set_difference}")

set_1 = {1, 2, 3, 4, 5, 6} # Reset

# Difference second method
set_1.difference_update(set_2)
print(f"Difference by changing the first set:\n{set_1}")

set_1 = {1, 2, 3, 4, 5, 6} # Reset

# Symmetric difference first method
set_symdifference = set_1.symmetric_difference(set_2)
print(f"Symmetric difference without changing first set:\n{set_symdifference}")

set_1 = {1, 2, 3, 4, 5, 6} # Reset

# Symmetric difference second method
set_1.symmetric_difference_update(set_2)
print(f"Symmetric difference by changing the first set:\n{set_1}")
````
